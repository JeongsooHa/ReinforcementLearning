```python
import numpy as np
def compute_CDF(arr, num_bins=1000):
    """
    usage: x, y = compute_CDF(arr):
           plt.plot(x, y)
    """
    values, base = np.histogram(arr, bins=num_bins)
    cumulative = np.cumsum(values)
    return base[:-1], cumulative / float(cumulative[-1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

path_dir = os.getcwd()
file_list = os.listdir(path_dir)
file_list.sort()

csv_file_list = []
rawData = []
for filename in file_list:
    if '.csv' in filename:
        csv_file_list.append(filename)
        rawData.append(pd.read_csv(filename))
        print(filename)
    
# print(rawData[0]['0'])
fig = plt.figure()

for i, data in enumerate(rawData):
    x, y = compute_CDF(data['0'].values.tolist())
    plt.plot(x, y)

    plt.xlabel('time')
    plt.ylabel('CDF')
#     plt.xlim(100,105)
#     plt.legend(csv_file_list[i][:-4], loc='upper center', bbox_to_anchor=(0.5, 1.05),
#           ncol=3, fancybox=True, shadow=True)
#     plt.legend(csv_file_list[i][:-4])
    
    print(csv_file_list[i][:-4])
L = plt.legend(csv_file_list[i][:-4])

for i,name in enumerate(csv_file_list):
    L.get_texts()[i].set_text(name[:-4])
plt.grid()
fig.savefig(os.getcwd()+'/total_.png')
# plt.close(fig)
plt
```
