```python
import numpy as np
import torch
np.set_printoptions(precision=3)
```

```python
mat = np.random.random((10,10))
# mat = torch.tensor(dist, dtype=torch.float)
```

```python
mask = np.zeros((10,10))
for i in range(8): 
    for ii in range(7):
        mask[i][ii] = 1
# mask = torch.tensor(mask, dtype=torch.float)
```

```python
mat, mask
```

```python
new_mat = np.multiply(mat, mask)
new_mat
```

```python
new_mat_t = torch.tensor(new_mat, dtype=torch.float)
```

```python
dist_t = torch.exp(new_mat_t)
```

```python
dist_t
```

```python
new_dist_t = torch.exp(dist_t)
```

```python
new_dist_t[dist_t==1] = 0
```

```python
dist_t, new_dist_t
```

```python
import sys
new2_dist_t = torch.exp(dist_t)
new2_dist_t[dist_t==1] = sys.maxsize
```
