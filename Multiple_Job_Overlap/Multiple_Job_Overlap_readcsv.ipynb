```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Job overlap
path_dir = os.getcwd()
file_list = os.listdir(path_dir)
file_list.sort()

csv_files = []
for csv_file in file_list:
    if '.csv' in csv_file:
        csv_files.append(csv_file)
# print(csv_files)

for filename in csv_files:
    df = pd.read_csv(path_dir+'/'+filename)
    for i in range(12):
        df = df.drop('n'+str(i),axis = 1)
        
    for i, count in enumerate(df.isna().sum()):
        if len(df['time']) == count:
            df = df.drop(str(i-2), axis = 1)

        
    # for i in range(len(csv_pd_y_data)):
    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(1,1,1)
    for i in range(len(df.columns)-2):
        ax.bar(df['time'], df[str(i)],alpha = 0.3, label=str(i))

    fig.suptitle("Number of remaining tasks for each task_"+filename[:-4], fontsize = 30)
    plt.xlabel('Env Time', fontsize=20)
    plt.ylabel('Number of tasks', fontsize=20)
    plt.xlim(0,1000)
    plt.ylim(0,30)
    ax.grid()
    ax.legend()

    fig.savefig(os.getcwd()+'/'+filename[:-4]+'.png')
    print("SAVED!" + filename[:-4])
    plt
    print("SHOWED! " + filename[:-4])
```
