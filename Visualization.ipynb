```python
import os
import pandas as pd
import numpy as np

csv_path_dir = '/Users/jeongsooha/Downloads/csv/HyperParams/'
category_list = os.listdir(csv_path_dir)
category_list.sort()
for category_name in category_list:
    if '.' is category_name[0]:
        continue
    new_csv_path_dir = csv_path_dir + category_name+"/"
    in_category_name = os.listdir(new_csv_path_dir)
    in_category_name.sort()
    
    for csv_folders in in_category_name:
        if '.' is csv_folders[0]:
            continue
        csv_folders_path_dir = new_csv_path_dir + csv_folders+"/"
        csv_files_list = os.listdir(csv_folders_path_dir)
        csv_files_list.sort()
        csv_graph = []

        for csv_files in csv_files_list:
            if '.' is csv_files[0] :
                continue
            csv_file = pd.read_csv(csv_folders_path_dir + csv_files)
            csv_graph.append(pd.DataFrame({csv_files[:-4]:csv_file[csv_file.columns[1]]}))
            
        cv = pd.merge(csv_graph[0], csv_graph[1], how='outer', left_index=True, right_index=True)
        for i in range(len(csv_graph)-2):
            cv = pd.merge(cv, csv_graph[i+2], how='outer', left_index=True, right_index=True)
        ax = cv.plot.line()
        title = csv_folders.replace("_"," ")
        ax.set_title(title)
        if "Ratio" in title:
            ax.set_ylim(0.8,1.1)
#         ax.set_xticks(csv_graph.index[0::50])
#         ax.set_xticklabels(csv_graph.index[0::10],rotation = 45, size=5)
#         ax.set_ylabel("Elapsed")
#         ax.set_xlabel("Iteration")
        ax.legend()
        ax.grid()
        ax
        fig = ax.get_figure()
        fig.savefig('/Users/jeongsooha/Downloads/csv/'+csv_folders+".png")
            
            
```
